# Scatterplots: Volume vs. Cost

```python
f,ax=plt.subplots(figsize=(17,10))
plt.scatter(df_item['Bottle_Volume_ml']/1000, df_item['Bottle_Cost'], alpha = 0.5)
ax.set_title('Volume vs. Cost-No limits')
plt.xlabel('Volume (liter)')
plt.ylabel('Cost')
plt.show()
```

```python
f,ax=plt.subplots(figsize=(17,10))
plt.scatter(df_item['Bottle_Volume_ml']/1000, df_item['Bottle_Cost'], alpha = 0.35)
# sns.swarmplot(df_item['Bottle_Volume_ml'], df_item['Bottle_Cost'])
ax.set_title('Volume vs. Cost')
plt.xlabel('Volume (liter)')
plt.ylabel('Cost')
plt.xlim([0.25,6.200])
plt.ylim([0.5,70])
plt.show()
```

# Scatterplot: Cost vs. Retail Price

```python
f,ax=plt.subplots(figsize=(17,10))
plt.scatter(df_item['Bottle_Cost'], df_item['Bottle_Retail_Price'], alpha = 0.35)
ax.set_title('Cost vs. Retail Price--No Limits')
ax.set_ylim(ymin = 0)
ax.set_xlim(xmin = 0)
plt.xlabel('Cost')
plt.ylabel('Retail Price')
plt.show()
```

```python
f,ax=plt.subplots(figsize=(17,10))
plt.scatter(df_item['Bottle_Cost'], df_item['Bottle_Retail_Price'], alpha = 0.25)
ax.set_title('Cost vs. Retail Price')
plt.xlabel('Cost')
plt.ylabel('Retail Price')
ax.set_ylim(ymin = 0)
ax.set_xlim(xmin = 0)
plt.xlim([1,100])
plt.ylim([1,150])
plt.show()
```
